APPLY OUR FACE DATASET

```python
import os
import cv2

# Fungsi untuk mengambil gambar wajah dan menyimpannya ke dalam folder Dataset
def capture_faces():
    # Membuat folder dataset\Phillia jika belum ada
    dataset_folder = 'dataset\Phillia'
    if not os.path.exists(dataset_folder):
        os.makedirs(dataset_folder)

    # Menggunakan kamera laptop (index 0)
    cap = cv2.VideoCapture(0)

    # Memastikan kamera terbuka
    if not cap.isOpened():
        print("Error: Unable to open camera.")
        return

    # Membuat detektor wajah
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Penghitung untuk nama file gambar
    count = 0

    while True:
        # Membaca frame dari kamera
        ret, frame = cap.read()

        if not ret:
            print("Error: Unable to read frame.")
            break

        # Mendeteksi wajah dalam frame
        faces = face_cascade.detectMultiScale(frame, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

        # Jika wajah terdeteksi
        if len(faces) > 0:
            for (x, y, w, h) in faces:
                # Mengambil gambar wajah dengan ukuran 250x250
                face_img = frame[y:y+h, x:x+w]
                face_img = cv2.resize(face_img, (250, 250))

                # Menyimpan gambar wajah ke dalam folder dataset\Phillia
                file_path = os.path.join(dataset_folder, f'face_{count}.jpg')
                cv2.imwrite(file_path, face_img)

                print(f"Captured face {count + 1} saved at:", file_path)

                count += 1

        # Menampilkan frame dari kamera
        cv2.imshow('Camera', frame)

        # Jika tombol 'q' ditekan, keluar dari loop
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Menutup kamera dan jendela OpenCV
    cap.release()
    cv2.destroyAllWindows()

# Memanggil fungsi untuk mengambil gambar wajah dan menyimpannya ke dalam folder Dataset
capture_faces()
```

CAPTURE FRAME FROM VIDEO

```python
import cv2

def capture_frames(video_path, output_folder, num_frames):
    # Buka video
    cap = cv2.VideoCapture(video_path)

    # Periksa apakah video berhasil dibuka
    if not cap.isOpened():
        print("Error: Video tidak dapat dibuka")
        return

    # Hitung jumlah frame dalam video
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Tentukan interval untuk menangkap frame
    frame_interval = total_frames // num_frames

    # Baca dan simpan frame sesuai dengan interval yang ditentukan
    for i in range(0, total_frames, frame_interval):
        # Pindah ke frame ke-i
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)

        # Baca frame
        ret, frame = cap.read()

        # Periksa apakah bacaan frame berhasil
        if not ret:
            print("Error: Gagal membaca frame")
            break

        # Simpan frame
        cv2.imwrite(f"{output_folder}/frame_{i}.jpg", frame)

    # Tutup video
    cap.release()
    cv2.destroyAllWindows()

def main():
    # Path video input
    video_path = "D:\\Project_Face_Recognition 2\\video\\Tirtayuda\\WIN_20240402_18_34_28_Pro.mp4"

    # Folder output untuk menyimpan frame
    output_folder = "D:\\Project_Face_Recognition\\dataset\\Tirtayuda"

    # Jumlah frame yang ingin ditangkap
    num_frames = int(input("Masukkan jumlah frame yang ingin ditangkap: "))

    # Tangkap frame
    capture_frames(video_path, output_folder, num_frames)

if __name__ == "__main__":
    main()
```

RESIZE TO "X" PIXELS

```python
import os
import cv2

def resize_images(input_folder, output_folder, target_width=250, target_height=250):
    # Membuat folder output jika belum ada
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Mendapatkan daftar file dalam folder input
    files = os.listdir(input_folder)

    # Mengubah ukuran setiap gambar dan menyimpannya di folder output
    for file in files:
        # Baca gambar
        img = cv2.imread(os.path.join(input_folder, file))

        # Periksa apakah file yang dibaca adalah gambar
        if img is not None:
            # Mengubah ukuran gambar
            resized_img = cv2.resize(img, (target_width, target_height))

            # Menyimpan gambar yang telah diubah ukurannya
            cv2.imwrite(os.path.join(output_folder, file), resized_img)
        else:
            print(f"File {file} bukan gambar.")

def main():
    input_folder = "D:\Project_Face_Recognition 2\Folder simpan sementara\Tirtayuda"  # Folder input berisi gambar-gambar yang ingin diubah ukurannya
    output_folder = "D:\Project_Face_Recognition 2\dataset\Tirtayuda"  # Folder output untuk menyimpan gambar-gambar yang telah diubah ukurannya

    # Mengubah ukuran gambar dalam folder input dan menyimpannya di folder output
    resize_images(input_folder, output_folder)

if __name__ == "__main__":
    main()
```
